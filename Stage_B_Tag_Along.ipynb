Running Penalization

```python
#define the target and non target learning variables

x = df.drop(['lights', 'date'], axis =1)

# y = df['Appliances'] # target variable (dependable vairable) as y

x.head()
```

```python
#Firstly, we normalise our dataset to a common scale using the min max scaler

from sklearn.preprocessing import MinMaxScaler

from sklearn.linear_model import LinearRegression
scaler = MinMaxScaler()

normalised_df = pd.DataFrame(scaler.fit_transform(x), columns=x.columns)

features_df = normalised_df.drop(columns=['Appliances'])

Appliances = normalised_df['Appliances']

#Now, we split our dataset into the training and testing dataset. Recall that we

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features_df, Appliances,
test_size=0.3, random_state=1)

linear_model = LinearRegression()

#fit the model to the training dataset
linear_model.fit(x_train, y_train)

#obtain predictions
predicted_values = linear_model.predict(x_test)

#MAE
from sklearn.metrics import mean_absolute_error, mean_squared_error
mse = mean_squared_error(y_test, predicted_values)
round(mse, 3)
```

```python
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=0.5)
ridge_reg.fit(x_train, y_train)
```

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(x_train, y_train)
```

```python
##comparing the effects of regularisation

def get_weights_df(model, feat, col_name):
    #this function returns the weight of every feature
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)

    return weights_df

lasso_weights_df = get_weights_df(lasso_reg, x_train, 'Lasso_weight')
lasso_weights_df
```

```python
lasso_predict = lasso_reg.predict(x_test)
```

```python
# Evaluating LASSO Regression

mse_lasso = mean_squared_error(y_test, lasso_predict)
round(mse_lasso, 3)
```
